```python
from scipy.signal import butter, filtfilt

# Define filter
def bandpass_filter(signal, lowcut=0.5, highcut=50.0, fs=record.fs, order=2):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype="band")
    return filtfilt(b, a, signal)

# Apply filter
filtered_signal = bandpass_filter(record.p_signal[:, 0])

# Plot filtered signal
plt.figure(figsize=(12, 4))
plt.plot(filtered_signal, color='purple', label="Filtered ECG")
plt.title("Filtered ECG Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

```python
import wfdb
import numpy as np
record = wfdb.rdrecord('mitdb/100')
annotations = wfdb.rdann('mitdb/100', 'atr')
ecg_signal = record.p_signal[:, 0] 
annotation_symbols = annotations.symbol
annotation_samples = annotations.sample
ecg_signal = (ecg_signal - np.mean(ecg_signal)) / np.std(ecg_signal)
```

```python
heartbeats = []
labels = []

for i in range(len(annotation_samples)):
    start = annotation_samples[i] - 100 
    end = annotation_samples[i] + 100    
    heartbeat = ecg_signal[start:end]
    if len(heartbeat) == 200:  
        heartbeats.append(heartbeat)
        labels.append(annotation_symbols[i])

heartbeats = np.array(heartbeats)
labels = np.array(labels)
```

```python
from scipy.stats import skew, kurtosis

features = []
for heartbeat in heartbeats:
    mean = np.mean(heartbeat)
    std = np.std(heartbeat)
    skewness = skew(heartbeat)
    kurt = kurtosis(heartbeat)
    features.append([mean, std, skewness, kurt])

features = np.array(features)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Testing labels shape: {y_test.shape}")
```

```python
from collections import Counter
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# Step 1: Filter out the 'V' class (1 sample) from training data
mask = (y_train != 'V')  # Exclude 'V' class
X_train_filtered = X_train[mask]
y_train_filtered = y_train[mask]

# Step 2: Apply SMOTE to balance the 'A' and 'N' classes
smote = SMOTE(random_state=42, k_neighbors=1)

# Reshape for SMOTE (flatten if necessary)
X_train_reshaped = X_train_filtered.reshape(X_train_filtered.shape[0], -1)

# Apply SMOTE
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_reshaped, y_train_filtered)

# Step 3: Encode the labels
label_encoder = LabelEncoder()
y_train_balanced_encoded = label_encoder.fit_transform(y_train_balanced)

# Step 4: Print class distribution after filtering and SMOTE
print("Class distribution after filtering 'V':", Counter(y_train_filtered))
print("Class distribution after SMOTE:", Counter(y_train_balanced))

# Optionally, check the shape of the reshaped data
print("Shape of X_train_balanced:", X_train_balanced.shape)
print("Shape of y_train_balanced:", y_train_balanced.shape)


```

```python
from scipy.signal import find_peaks


r_peaks, _ = find_peaks(ecg_signal, distance=200)  

plt.plot(ecg_signal)
plt.scatter(r_peaks, ecg_signal[r_peaks], color='red', label='R-peaks')
plt.title('ECG Signal with R-peaks')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python

heartbeats = []
labels = []
window_size = 200  
for i in range(len(r_peaks)):
    start = r_peaks[i] - 100  
    end = r_peaks[i] + 100   
    heartbeat = ecg_signal[start:end]
    if len(heartbeat) == window_size:
        heartbeats.append(heartbeat)
        labels.append(annotation_symbols[i])
heartbeats = np.array(heartbeats)
labels = np.array(labels)
```
